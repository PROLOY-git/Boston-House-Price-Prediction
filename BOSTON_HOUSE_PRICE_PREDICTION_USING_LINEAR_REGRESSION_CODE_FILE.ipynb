# ML models need data in array format for one Dimension- Only X

```python
X = np.array(x).reshape(-1,1)### The -1 in reshape(-1, 1) tells NumPy to automatically calculate the number of rows needed based on the total number of elements in the array
```

```python
X
```

```python
Y = np.array(y).reshape(-1,1)  ###The -1 in reshape(-1, 1) tells NumPy to automatically calculate the number of rows needed based on the total number of elements in the array
```

```python
Y
```

# Spliting the Data with 80% and 20% Ratio between X and Y

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(X,Y, random_state=5, test_size = 0.20, train_size=0.80 )
```

```python
X_train.mean()
```

```python
X_train.shape
```

```python
X_val.shape
```

```python
Y_train.shape
```

```python
Y_val.shape
```

```python
slrm = LinearRegression()
```

```python
slrm
```

```python
slrm.fit(X_train, Y_train)
```

```python
y_pred = slrm.predict(X_val)
```

```python
y_pred
```

```python
Y_val
```

```python
mse = mean_squared_error(y_pred, Y_val)
```

```python
mse
```

```python
rmse = np.sqrt(mse)
```

```python
rmse
```

```python
y.min()
```

```python
y.max()
```

```python
y.mean()
```

```python
error = (rmse/y.mean())*100
```

```python
error
```

```python
Accuracy = 100 - error
```

```python
Accuracy
```
